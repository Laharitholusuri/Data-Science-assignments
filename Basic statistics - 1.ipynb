## Descriptive Analytics for Numerical Columns

```python
import pandas as pd

# reading the csv file and displaying the data 
df = pd.read_csv("sales_data_with_discounts.csv");
df
```

```python
df.info()
```

```python
# selecting numerical columns from the dataset.

numeric_columns = df.select_dtypes(include=['int64', 'float64'])
numeric_columns
# i used .select_dtypes() method to return only the columns that have numeric data types such as int64 or float64.


# we can also select numeric columns from the dataframe using .iloc() and pd.concat() from below code:
# .iloc[] selects columns based on index positions and pd.concat() is used to combine columns using axis=1

# num1 = df.iloc[:,8:13]
# num2 = df['Volume']
# numeric_columns = pd.concat([num2,num1],axis=1)
# numeric_columns.head()
```

```python
mean = numeric_columns.mean()
print(mean)
# .mean() used to calculates the arithmetic average of each numerical column.
```

```python
median = numeric_columns.median()
print(median)
# .median() used to finds the middle value of each column.
```

```python
numeric_columns.mode().iloc[0]
# .mode() used to return the most common value's. I use .iloc[0] to show the first mode.
```

```python
numeric_columns.std()
# .std() used to measures data spread (higher value means more variation).
```

```python
numeric_columns.describe()
```

### Brief interpretation:
#### Most numerical columns (Avg Price, Total Sales Value, Discount Amount, Net Sales Value) shows mean much higher than median, meaning they are positively skewed.
#### High standard deviations indicate large variation in prices, sales values, and discounts.
#### Volume has low variation and is almost symmetrical.
#### Discount Rate (%) is very stable with low std discounts are consistent.

## Histograms

```python
import matplotlib.pyplot as plt

# Plotting histogram for each numerical columns
for col in numeric_columns:
    plt.figure(figsize=(6,4))
    
    plt.hist(df[col], bins=20)
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    
    plt.show()

# matplotlib library is used for data visualization.
# histogram shows data distribution for each numerical columns and helps us see the skewness (left or right) and presence of outliers.
```

## Box Plots

```python
import seaborn as sns
for col in numeric_columns:
    sns.boxplot(x=df[col], color='lightblue')
    plt.title(col)
    plt.show()

# seaborn library is used for better data visualizations
# Boxplot is used to visualize outliers and interquartile range.
# Boxplots show outliers as dots and helps to identify the spread (IQR) of the data.
```

```python
# Bar Chart Analysis for Categorical Column: 
# Identifying categorical columns

categ_columns = df.select_dtypes(include=['object'])
print(categ_columns.head())

# i used .select_dtypes() method to return only the columns that have categorical data types such as object.
```

## Bar Charts

```python
for col in categ_columns:
    df[col].value_counts().plot(kind='bar', color='orange', edgecolor='black')
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()

# Bar charts shows frequency of each category
```

### Histograms:
#### Most numerical data is positively skewed — few items contribute to high sales and prices:
#### Avg Price, Total Sales Value, Discount Amount, Net Sales Value are right-skewed due to a few very high values.
#### Volume shows slight right skew because some transactions have much higher quantities.
#### Most numeric columns show long right tails in histograms, confirming skewness.

### Box Plots: 
#### Presence of outliers in sales and price columns — representing top-performing or high-value products:
#### Avg Price, Total Sales Value, Discount Amount, and Net Sales Value contain multiple outliers.
#### These outliers represent expensive products or high-value sales.
#### Volume shows only a few outliers and appears more consistent.
#### Discount Rate (%) has almost no outliers, showing uniform discount distribution.

### Bar Charts: 
#### A small number of brands and models dominate, showing concentration of sales among top performers:
#### The dataset mainly belongs to one business unit (Mobiles).
#### Some brands appear more frequently, showing uneven brand distribution.
#### The Model column has many unique models, indicating high product variety.
#### Category counts are imbalanced, meaning some categories dominate the dataset.
#### City and Day distribution may also be uneven depending on sales spread.
