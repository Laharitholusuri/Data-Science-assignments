## Exercise 1: Prime Numbers

```python

# A prime number is a natural number greater than 1 that has no positive divisors other than 1 and itself.

Num = int(input("Enter a Number:"))
if Num > 1:
    for i in range(2, Num):
        if Num % i == 0:
            print(Num, "is not a prime number")
            break
    else:
        print(Num, "is a prime number")
else:
    print(Num, "is not a prime number")

# I gave input Num = 7. Firstly it checks for 7 > 1, then for loop iterates from 2 to 6 to check if 7 is divisible by any of these numbers.
# Since 7 is not divisible by any those numbers except 1 and itself, the program prints "7 is a prime number".
```

## Exercise 2: Product of Random Numbers

## Exercise 3: Squares of Even/Odd Numbers

```python

for i in range(100, 200):
    if i % 2 == 0:
        print(i, ":", i ** 2)

# i choose to print all even numbers between 100 and 200 with their squares.
# The for loop iterates from 100 to 200 and checks if the number is divisible by 2 or not.
# if it is divisible, then it prints square of the number.
```
